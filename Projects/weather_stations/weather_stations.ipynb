```python
from selenium import webdriver
import time
import pandas as pd
import math
```

```python
titles = []
station_ids = []
counties_all = []

# options = webdriver.ChromeOptions()
# options.add_argument('headless')
browser = webdriver.Chrome('./chromedriver')
browser.get('https://www.ncdc.noaa.gov/cdo-web/datatools/selectlocation')
browser.maximize_window()

browser.find_element_by_xpath('//*[@id="datasetSelect"]/option[1]').click()
time.sleep(1)
browser.find_element_by_xpath('//*[@id="locationCategorySelect"]/option[4]').click()
time.sleep(3)
states = browser.find_elements_by_xpath('//*[@id="selectedCountyState"]/option')
```

```python
for state in states:
    time.sleep(3)
    state.click()
    time.sleep(1)
    counties = browser.find_elements_by_xpath('//*[@id="selectedCounty"]/option')
    for county in counties:
        time.sleep(1)
        county.click()
        time.sleep(4)
        count = browser.find_element_by_xpath('//*[@id="locationPagination"]/li[2]/span[3]').text
        if count != '':
            pages = math.ceil(int(count) / 25)
        else:
            print('County, State: ', county, state, 'No count found')
            pages = 1
        print('County: ', county.text)
        print('State: ', state.text)
        print('Count: ', count)
        for page in range(pages):
            time.sleep(3)
            stations = browser.find_elements_by_xpath('//*[@id="stationsTable"]/tbody/tr')
            for station in stations:
                counties_all.append(county.text)
                try:
                    title = station.find_element_by_class_name('title').text
                    titles.append(title)
                except:
                    print('No title')
                    titles.append(None)
                try:
                    id_ = station.find_element_by_class_name('stationID').text
                    station_ids.append(id_.split(':')[1])
                except:
                    print('No id')
                    station_ids.append(None)
            print('Page number: ', page)
            if page + 1 < pages:
                browser.find_element_by_xpath('//*[@id="locationPagination"]/li[3]').click()
        print('-' * 100)
```

```python
pd_dict = dict()
pd_dict['County'] = counties_all
pd_dict['Station Name'] = titles
pd_dict['Station Id'] = station_ids

df = pd.DataFrame(pd_dict)

df.to_excel('weather_stations.xlsx', index=False)
```
