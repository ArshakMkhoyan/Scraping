```python
import sys
sys.path.append('../')
from Base import Scraper

import pandas as pd
import time
```

```python
scraper = Scraper()
scraper.navigate('https://www.manta.com/search?search_source=nav&search=Restaurants&search_location=Los+Angeles+CA&pt=34.0396%2C-118.2661')

## Number of pages
ints = []
scraper.driver.find_element_by_xpath("/html/body/div[2]/div[2]/div[3]/div/div/div/ul/li[9]/a").click()
time.sleep(1)
for i in scraper.driver.find_elements_by_xpath("//ul[@class='pagination pagination-md mll']//li"):
    try:
        ints.append(int(i.text))
    except:
        pass
n_pages = ints[-1]
print('Number of pages:', n_pages)

names = []
link_to_rest = []
manta_verified = []
streetAddress = []
addressLocality = []
addressRegion = []
postalCode = []
telephone = []
Website = []
description = []

## Scrap each page
for n_page in range(1, n_pages + 1):
    link = f'https://www.manta.com/search?search=Restaurants&pg={n_page}&pt=34.0396,-118.2661&search_location=Los+Angeles+CA&pt=34.0396%2C-118.2661'
    scraper.navigate(link)
    rests = scraper.driver.find_elements_by_xpath("/html/body/div[2]/div[2]/div[2]/div[3]/div/ul/li")
    
    print('Page number:', n_page)
    
    for rest in rests:
        try:
            print(rest.find_element_by_xpath(".//strong[@itemprop='name']").text)
            names.append(rest.find_element_by_xpath(".//strong[@itemprop='name']").text)
        except:
            print('No Name')
            names.append(None)
        try:
            link_to_rest.append(rest.find_element_by_xpath(".//a[@class='media-heading text-primary h4']").get_attribute('href'))
        except:
            print('No link to restaurant')
            link_to_rest.append(None)
        try:
            rest.find_element_by_xpath(".//div[@title='Manta Verified']")
            manta_verified.append(True)
        except:
            print('No manta verification')
            manta_verified.append(False)
        try:    
            streetAddress.append(rest.find_element_by_xpath(".//span[@itemprop='streetAddress']").text)
        except:
            print('No street address')
            streetAddress.append(None)
        try:
            addressLocality.append(rest.find_element_by_xpath(".//span[@itemprop='addressLocality']").text)
        except:
            print('No city')
            addressLocality.append(None)
        try:
            addressRegion.append(rest.find_element_by_xpath(".//span[@itemprop='addressRegion']").text)
        except:
            print('No region')
            addressRegion.append(None)
        try:    
            postalCode.append(rest.find_element_by_xpath(".//span[@itemprop='postalCode']").text)
        except:
            print('No postal code')
            postalCode.append(None)
        try:
            telephone.append(rest.find_element_by_xpath(".//div[@itemprop='telephone']").text)
        except:
            print('No telephone')
            telephone.append(None)
        try:
            Website.append(rest.find_element_by_xpath(".//*[contains(text(), 'Website')]").get_attribute('href'))
        except:
            print('No website')
            Website.append(None)
        try:    
            description.append(rest.find_element_by_xpath(".//div[@itemprop='description']").text)
        except:
            print('No description')
            description.append(None)
        print('-' * 100)
    print('_' * 100)
```

```python
pd_dict = dict()
pd_dict['Restaurant name'] = names
pd_dict['Link in manta'] = link_to_rest
pd_dict['Manta verified'] = manta_verified
pd_dict['Street address'] = streetAddress
pd_dict['City'] = addressLocality
pd_dict['State'] = addressRegion
pd_dict['Postal code'] = postalCode
pd_dict['Phone number'] = telephone
pd_dict['Website'] = Website
pd_dict['Description'] = description


df = pd.DataFrame(pd_dict)
df.to_csv('manta_LA_restaurants.csv', index=False)
```
