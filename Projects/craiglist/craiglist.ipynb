```python
scraper = Scraper()
scraper.navigate('https://bakersfield.craigslist.org/')
```

## Find all categories

```python
categories = scraper.driver.find_elements_by_class_name('col')

for category in categories:
    name_category = category.find_element_by_class_name('ban')
    print(name_category.text, name_category.find_element_by_partial_link_text('').get_attribute("href"))
    for good in category.find_elements_by_xpath(".//div[@class='cats']//ul//li"):
        print(good.text, good.find_element_by_xpath('.//a').get_attribute("href"))
    print('-' * 100)
```
