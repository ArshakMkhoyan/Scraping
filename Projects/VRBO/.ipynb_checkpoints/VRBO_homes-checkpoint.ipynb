```python
import requests
from selenium import webdriver
from scrapy.http import TextResponse
import time
import pandas as pd
from collections import defaultdict
```

```python
def find_name(text):
    for t in text:
        if 'Property' in t:
            return t.split(':')[1].strip()
    return None

def find_sleeps(text):
    for point in text.split('\n'):
        if 'Sleeps' in point:
            return point.split(':')[1].strip()
```

```python
links = []
prices = []
names = []
descriptions = []
body_titles = []
sleeps = []

options = webdriver.ChromeOptions()
options.add_argument('headless')
browser = webdriver.Chrome('../../chromedriver', options=options)

for page_n in range(1, 7):
    links_tmp = []
    print('Page number:', page_n)
    link = f'https://www.vrbo.com/search/keywords:saint-petersburg-fl-usa/page:{page_n}/arrival:2020-03-06/departure:2020-03-07?petIncluded=false&ssr=true&adultsCount=1'
    browser.get(link)
    time.sleep(1)

    y = 600
    for timer in range(0,25):
        browser.execute_script("window.scrollTo(0, "+str(y)+")")
        y += 600  
        time.sleep(2)
        
    homes = browser.find_elements_by_xpath("//div[@data-wdio='ScrollListener']//div[@class='HitCollection']/div")
    print('Number of homes:', len(homes))
    
    for home in homes:
        text = home.text.split('\n')
        names.append(find_name(text))
        links_tmp.append(home.find_elements_by_partial_link_text('')[1].get_attribute("href"))
        if names[-1] == None:
            print('Not found name')
            print(names[-1])
            print(text)
    
    for link in links_tmp:
        if link == None:
            prices.append(None)
            descriptions.append(None)
            body_titles.append(None)
            sleeps.append(None)
            print('Doesnt have a link')
            continue
        browser.get(link)
        browser.maximize_window()
        time.sleep(2)
        
        try:
            browser.find_element_by_class_name("MoreOrLess__toggle-text").click()
        except:
            pass
        y = 400
        for timer in range(0,8):
            browser.execute_script("window.scrollTo(0, "+str(y)+")")
            y += 400  
            time.sleep(1)
        try:
            prices.append(browser.find_element_by_class_name("rental-price__amount").text[1:])
        except:
            print('Price not found')
            print(link)
            prices.append(None)
        try:
            sleeps.append(find_sleeps(browser.find_element_by_class_name("listing-bullets").text))
        except:
            sleeps.append(None)
            print('Number of sleeps not found')
            print(link)
        try:
            body_title = browser.find_element_by_class_name("collapsible-content__body-title").text
            body_titles.append(body_title)
        except:
            print('Title not found')
            print(link)
            body_titles.append(None)

        description_list = browser.find_elements_by_xpath("//div[@class='collapsible-content__body-text u-freetext-fix']//p")
        if description_list != []:
            description = '\n'.join([i.text for i in description_list])
            descriptions.append(description)
        else:
            print('Body not found')
            print(link)
            descriptions.append(None)
    links.extend(links_tmp)
```

```python
pd_dict = defaultdict(list)
pd_dict['Headline'] = names
pd_dict['Price ($/day)'] = prices
pd_dict['Sleeps'] = sleeps
pd_dict['Links'] = links
pd_dict['Description Title'] = body_titles
pd_dict['Description'] = ['\n'.join([k for k in i.replace('<br/>', '\n').split('\n') if k != '']).strip() if type(i) == str else None for i in descriptions] 

df = pd.DataFrame(pd_dict)

df.to_excel('vrbo_data_st_petersburg_sample.xlsx', index=False)
```
