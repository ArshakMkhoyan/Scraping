```python
import requests
from selenium import webdriver
from scrapy.http import TextResponse
import time
import pandas as pd
from collections import defaultdict
```

```python
browser = webdriver.Chrome('../../chromedriver')
browser.get('https://www.startengine.com/explore')
browser.maximize_window()

## Upload all companies
while True:
    try:
        browser.find_element_by_xpath("//div[@class='text-center margin-bottom-standard']//a").click()
    except:
        break
        
all_companies = browser.find_elements_by_xpath("//div[@class='all-tombstones clearfix']/div")

# Get all company names and links
company_names = []
company_links = []
for company in all_companies:
    try:
        company_links.append(company.find_element_by_partial_link_text('').get_attribute("href"))
    except:
        print('link not found')
        company_links.append(None)
    try:
        company_names.append(company.find_element_by_tag_name('h5').text)
    except:
        print('name not found')
        company_names.append(None)
        
# Data for each company
company_coments = defaultdict(list)
for n, comp_link in enumerate(company_links):
    names = []
    investments = []
    invested = []
    days_ago = []
    user_questions = []
    reply_body = []
    
    browser.get(comp_link)
    y = 1000
    for timer in range(0,5):
        browser.execute_script("window.scrollTo(0, "+str(y)+")")
        y += 1000  
        time.sleep(0.2)
    
    # Upload all comments (scroll untill page end)
    while True:
        time.sleep(2)
        list_of_show_more = browser.find_elements_by_class_name('show-all')
        namings = []
        for i in list_of_show_more:
            namings.append(i.text)
        for k, name in enumerate(namings):
            if 'comments' in name.lower():
                list_of_show_more[k].click()
                break
        else:
            break
    
    comments_answers = browser.find_elements_by_xpath('//*[@id="profile-comments"]/div/div/ul/li')
    print('_' * 100)
    print(f'website number: {n}')
    print('Link', comp_link)
    print(f'Number of comments: {len(comments_answers)}')
    for comment_answer in comments_answers:
        # Get user data
        try:
            user_data = comment_answer.find_element_by_class_name('user-info')
            try:
                names.append(user_data.find_element_by_class_name('user-name').text)
            except:
                names.append(None)
            try:
                investments_data = user_data.find_elements_by_class_name('tag')
                inv_n = False
                inv_flag = False
                for inv in investments_data:
                    if 'investments' in inv.text.lower():
                        investments.append(inv.text.lower())
                        inv_n = True
                    if 'invested' in inv.text.lower():
                        invested.append(True)
                        inv_flag = True
                if not inv_n:
                    investments.append(None)
                if not inv_flag:
                    invested.append(None)
            except:
                investments.append(None)
                invested.append(False)
            try:
                days_ago.append(user_data.find_element_by_class_name('comment-date').text)
            except:
                days_ago.append(None)
        except:
            print('Not found User Data')
            names.append(None)
            investments.append(None)
            days_ago.append(None)
        # Get questions
        try:
            user_questions.append(comment_answer.find_element_by_class_name('clearfix').text)
        except:
            print('Not found question')
            user_questions.append(None)
        # Get Answers
        try:
            reply = comment_answer.find_element_by_class_name('margin-bottom-standard')
            try:
                reply_body.append(reply.find_element_by_class_name('comment-body').text)
            except:
                print('Not found answer')
                reply_body.append(None)
        except:
            print('Not found answer')
            reply_body.append(None)
        
    company_coments[(company_names[n], comp_link)] = [names, investments, invested, days_ago, user_questions, reply_body]
```

```python
final_company_name = []
final_company_link = []
final_names = []
final_investments = []
final_invested = []
final_days_ago = []
final_user_questions = []
final_reply_body = []
for company, data in company_coments.items():
    for i in range(len(data[0])):
        if data[-1][i] == None:
            continue
        final_company_name.append(company[0])
        final_company_link.append(company[1])
        final_names.append(data[0][i])
        final_investments.append(data[1][i].split()[0] if data[1][i] != None else None)
        final_invested.append('Yes' if data[2][i] else 'No')
        final_days_ago.append(data[3][i])
        final_user_questions.append(data[4][i])
        final_reply_body.append(data[5][i])

df = pd.DataFrame({'Company Name': final_company_name, 'Company Link': final_company_link, 'User Name': final_names,
             'Number of Investments': final_investments, 'Invested Flag': final_invested,
              'Amount if Time Ago': final_days_ago, 'Question': final_user_questions, 'Answer': final_reply_body})

df.to_csv('startengine.csv', index=False)
```
