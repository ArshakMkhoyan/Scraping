# 1) Scrape all the latest news headings (and publication time) from https://www.tert.am/en
and answer the following questions:<br>
- What are the top most frequent words? (not stopword),<br>
- How many typical Armenian surnames can be counted in titles? (assume surname in
any word starting with an uppercase and ending with “yan”),<br>
- In what part of the day are most of the articles published? (you may provide some
range such as from 10am to 11am).

```python
def clear_str(string):
    return string.replace(u'\xa0', u' ')

response = requests.get('https://www.tert.am/en')
page = TextResponse(url=response.url,
                   body=response.text,
                   encoding='utf-8')

date_topic_comb = page.css("ul[class='list list--with-scroll scroller-block show-link-visited']>li>div>a>span::text").extract()
date_topic_comb = [clear_str(string=string) for string in date_topic_comb]
date_topic_split = [[date_topic_comb[i], date_topic_comb[i+1]] for i in range(0, len(date_topic_comb), 2)]
print(*date_topic_split[:4], sep='\n')
```

```python
df = pd.DataFrame(data=date_topic_split, columns=['Time', 'Topic'])
df.head()
```

#### Cleaning text

```python
whole_text = TextBlob("".join(df.Topic.tolist()))
sw = stopwords.words('english')
words = whole_text.words.lemmatize().lower()

clean_text = [word for word in words if word not in sw and len(word) > 2]
clean_text[:4]
```

### Part A
What are the top most frequent words?

```python
freq = nltk.FreqDist(clean_text)
freq.plot(25)
```

### Part B
How many typical Armenian surnames can be counted in titles?

```python
pattern = '[A-Z][a-z]*[yi]an'
surnames = re.findall(pattern=pattern, string="".join(df.Topic.tolist()))
surnames = [surname for surname in set(surnames) if surname not in ['Armenian', 'Russian']]
print(f'Number of different Armenian surnames: {len(surnames)}')
```

### Part C
In what part of the day are most of the articles published?

```python
df.Time.apply(lambda x: x[:3]).value_counts()
```

# 2) Scrape all jobs and companies from https://staff.am/en/jobs and answer the following
questions:<br>
- Which company has the highest number of job postings?<br>
- How many postings does this company have?<br>
- Which company has the highest number of “developer” (or related) job postings?<br>

```python
url = 'https://staff.am/en/'
def get_jobs_and_comps(url):
    response = requests.get(url)
    page = TextResponse(url=response.url,
                       body=response.text,
                       encoding='utf-8')

    jobs = page.css("div[class='job-inner job-item-title']>p[class='font_bold']::text").extract()
    companies = page.css("div[class='job-inner job-item-title']>p[class='job_list_company_title']::text").extract()
    return jobs, companies

def get_all_pages(url, n_pages=5):
    jobs, companies = [], []
    for n_page in range(n_pages):
        new_url = url + f'jobs?page={n_page}&per-page=50'
        new_jobs, new_companies = get_jobs_and_comps(new_url)
        jobs.extend(new_jobs)
        companies.extend(new_companies)
    return jobs, companies

jobs, companies = get_all_pages(url=url, n_pages=10)
staff_df = pd.DataFrame({'Job Title':jobs, 'Company':companies})
print(staff_df.head())
print(staff_df.shape)
```

### Part A and B
Which company has the highest number of job postings?

```python
staff_df.Company.value_counts()[:10]
```

### Part C

```python
staff_df[staff_df['Job Title'].str.lower().str.contains("developer")]["Company"].value_counts()[:5]
```
